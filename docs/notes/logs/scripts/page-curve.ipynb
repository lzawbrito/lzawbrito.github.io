```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.serif'] = "Computer Modern Roman"
plt.rcParams['xtick.top'] = False
plt.rcParams['ytick.right'] = False
```

```python
fig, ax = plt.subplots(figsize=(3,2))

x = np.linspace(0, 2, 100)
ax.plot(x,  -np.abs(x - 1), color='black', alpha=0.6)
ax.grid()
ax.set_yticks([0, -1])
ax.set_yticklabels(['$\\frac{1}{2} \ln \\text{dim}\; \mathcal{H}$', '0'])
ax.set_xticks([0, 2])
ax.set_xticklabels(['0', '$L$'])

ax.spines[['right', 'top']].set_visible(False)

ax.set_title('$S(\mathcal{A})$')
# ax.get_yaxis().tick_bottom()

ax.set_xlabel('$i$')

plt.savefig('../2023/figs/december/page-curve.pdf')
```
