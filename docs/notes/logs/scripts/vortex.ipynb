```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Meshgrid 
aspect = 2.5
density = 0.5
range = 5
fig, ax = plt.subplots(figsize=(3, 3*aspect))
x, y = np.meshgrid(np.arange(-range, range, density), np.arange(-range*aspect, range*aspect, density))

def vortex(x, y, x0, y0): 
	return -(y - y0)/(np.sqrt((x-x0)**2 + (y-y0)**2))**(2), (x-x0)/(np.sqrt((x-x0)**2 + (y-y0)**2))**(2)

def normalize(x, y, f=lambda x: x): 
	return x / f(np.sqrt(x**2 + y**2)), y / f(np.sqrt(x**2 + y**2))

# Directional vectors 
v1 = vortex(x, y, 0, -5)
v2 = vortex(x, -y, 0, -5)
u = v1[0] + v2[0] 
v = v1[1] + v2[1] 

u, v = normalize(u, v)#, np.exp)

# Plotting Vector Field with QUIVER 
ax.quiver(x, y, u, v, color='black', alpha=0.7,
		  headwidth=3, 
		  headlength=5, 
		  scale=30) 
# Setting x, y boundary limits 
#  
ax.tick_params(length=0)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.axis('equal')
# disable grid 
plt.grid()
plt.savefig('../../../Documents/logs/2023/figs/vortex.pdf')
plt.show() 
```
