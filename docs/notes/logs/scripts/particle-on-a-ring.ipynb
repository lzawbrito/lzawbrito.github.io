# Particle on a ring with winding number
$$
\phi(t) = \frac{2\pi}{\beta}Q t + \sum_\ell \phi_\ell \exp\left(\frac{2\pi\ell}{\beta}t\right)
$$

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.serif'] = "Computer Modern Roman"

fig, ax = plt.subplots(figsize=(5,3)) 
t = np.linspace(0, 1, 100)

def phi(t, q, l): 
	return 2 * np.pi * q * t + np.sin(2 * np.pi * l * t)

qs = np.arange(0, 4)
for q in qs:
	ax.plot(t, phi(t, q, 2), alpha=0.6, label=f'$Q={q}$')
	ax.plot(t, phi(t, q, 0), alpha=0.3, linestyle='dashed', color='black',)

ax.set_yticks(2 * np.pi * qs)
ax.set_yticklabels(['$0$', '$2\pi$', '$4\pi$', '$6\pi$'])
ax.set_ylabel('$\\phi$')
ax.set_xlabel('$\\tau/\\beta$')
ax.legend(frameon=False)

# plt.savefig('../2023/figs/november/winding-number-particle-on-ring.pdf')
```

$$
\left(n+\frac{1}{2}\right)^2 
$$

```python
def e(k, th): 
	return 2 * np.power(k - th/(2*np.pi), 2)

range = 5

k = np.linspace(-range, range, 100)
n = np.arange(-range, range + 1)

fig, ax = plt.subplots(figsize=(4, 2))

th = np.arange(0, 3)

for t in th:
	ax.plot(k, e(k, np.pi*t),alpha=0.5,label=f'${t}\pi$')
	ax.scatter(n, e(n, np.pi*t))

ax.set_ylim(-1, 4)
ax.set_xlim(-1.5, 3)
ax.set_ylabel('$E_n$')
ax.set_xlabel('$n$')
ax.legend(frameon=False)

plt.savefig('../2023/figs/november/spectrum-particle-on-ring.pdf')
```
